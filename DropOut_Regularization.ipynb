This is for machine learning overfitting and underfitting problems.
Over Fitting: When you have training data and you try to train a model to much then it might over fit due to the occurance of same data again and again.
Data regularization is used to tackle over fitting problems in deep learning fields.

For Best Machine Learning practice just try to find the best fitted line.
The over fitting is also depend on epochs size. if you give more than 100 epochs value then the training model will be over fit.

## DropOut
If dropout=0.20 then out of 10 neurons 2 neurons will be dropped
In Each epochs it will take different neurons for dropping
If you select fropout=0.50 it means then from 10 neurons it will drop 5 neurons for training.

Neurons will not learn reduntant details of inputs.                     

```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df = pd.read_csv('sonar_dataset.csv', header=None) #Header=None will assign columns in a sequence from 0 till the last columns.
                                                    #In case if your column has no Name then use it.
df.sample(5)
```

```python
df.shape #208 rows and 61 columns
```

```python
df.columns #Here 60 is the target column for the data.
```

```python
df[60].value_counts() #It will shows that which value occurs how many times like m occurs 111 times out of 208.
```

```python
X = df.drop(60, axis='columns') #For splitting the data 
y = df[60] 
#Onehot Encoding
X
```

```python
#Setting the y column into 0 or 1
y = y.map({'R': 0, 'M': 1})
y.sample(5)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1) #We use random_state=1 b/c of reproducibility. for running again and again
```

```python
X_train.shape, X_test.shape 
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.activations import sigmoid,relu
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
#Input_shape depend of number of features columns. As we have 60 columns for feature data that's why i use 60 in input_shape
model = keras.Sequential([
    Dense(60, activation='relu', input_shape=(60,)), 
    Dense(30, activation='relu'),
    Dense(15, activation='relu'),
    Dense(1, activation='sigmoid') #this last one is for output that's why we use units=1 because we get just one out value of a one sample.
 ])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train, y_train, epochs=70, batch_size=8)
```

```python
model.evaluate(X_test, y_test) #the prediction is so it means that out of 100 values it predicted 78 values corrected.
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[: 10])

#Round it between 0 and 1
y_pred = np.round(y_pred)
print(y_pred[: 10])
```

```python
#Now trying our model with Dropout layer
#Input_shape depend of number of features columns. As we have 60 columns for feature data that's why i use 60 in input_shape
model1 = keras.Sequential([
    Dense(60, activation='relu', input_shape=(60,)), 
    Dropout(0.5), #best practice is to put dropout after each hidden layer
    Dense(30, activation='relu'),
    Dropout(0.5),
    Dense(15, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid') #this last one is for output that's why we use units=1 because we get just one out value of a one sample.
 ])

model1.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model1.fit(X_train, y_train, epochs=70, batch_size=8)
#Using dropout may loss your accuracy 
#But the performance is improved
```

```python
model1.evaluate(X_test, y_test)
```
