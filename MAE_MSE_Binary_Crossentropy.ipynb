```python
import numpy as np
```

```python
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python
def mae(y_true, y_predicted):
    total_error = 0
    for yt,yp in zip(y_true, y_predicted):
        total_error += abs(yt - yp)

    print("total error:", total_error)
    mae = total_error / len(y_true)
    return mae
```

```python
# mae(y_true, y_predicted) #this function is calculating mean absolute error. So we provide predicted and target values both
#Other way of doing it via Numpy functions
np.mean(np.abs(y_predicted - y_true)) #and the results are same. its very simple 
```

```python
epsilon = 1e-15
```

```python
y_predicted_new = [max(i,epsilon) for i in y_predicted] #this we are doing is just for log(0) B/C it's equal to undefine
y_predicted_new
```

```python
y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new] 
y_predicted_new
```

```python
y_predicted_new = np.array(y_predicted_new)
np.log(y_predicted_new)
```

```python
#formula for Binary_crossentropy from scratch
#its simple via numpy and we calculate log los or Binray_crossentropy and this is how we calculate it by reminding the formula in coding
-np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```
